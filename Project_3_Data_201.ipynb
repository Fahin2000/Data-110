**Project 3 - Exploring Excellence: Top 1000 Global Universities**

This dataset is related to the ranking of 1000 top universities in the world. It has fourteen columns and one thousand rows. The dataset is updated a year ago in Kaggle. It is including columns for world rank, institution, location, quality of education, score and many more. As education is an essential part of our lives, every student is very interested to know which educational institutions are doing well and what can be best for them.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os as os
```

```python
df = pd.read_csv('A Comprehensive Overview of the Worlds Top 1000 Universities a 14-column dataset.csv')
```

Data Visualization

```python
Quality_of_Education_column = 'Quality of Education'
value_column = 'Score'
```

```python
Top_10_Universities = df.head(10)
```

```python
plt.bar(Top_10_Universities[Quality_of_Education_column], Top_10_Universities[value_column], color='pink')
plt.xlabel('Quality of Education')
plt.ylabel('Score')
plt.title('Bar Graph of Top 1000 Universities')
plt.xticks(rotation = 45)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.relplot(x = 'National Rank',
            y = "Score",
            data = df,
            kind= "scatter",
            hue = 'Location',
            )

plt.title('Top 1000 Universities in the World')
plt.xlabel('National Rank')
plt.ylabel('Score')
plt.ylim(0,150)
plt.xlim(0,200)
plt.show()
```

```python
# Boxplot for Research Output by National Rank
plt.figure(figsize=(25, 20))
plt.rcParams.update({'font.size': 10})
plt.subplot(2, 1, 2)
sns.boxplot(data = df, x='Research Output', y= 'National Rank', hue= 'Location')

plt.title('Top 1000 Universities')
plt.xlabel('Research Output')
plt.ylabel('National Rank')
plt.legend(title='Location')
plt.ylim(0,100)
plt.show()

```
